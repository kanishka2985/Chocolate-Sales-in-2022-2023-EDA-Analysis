## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
df1 = df.copy()
df1.head()
````

````python
df1 = df1.dropna()
df1.shape
````

````python
# Forces Pandas to assume the day comes first (e.g., DD/MM/YYYY) when ambiguous
df1["Order_Date"] = pd.to_datetime(df1["Order_Date"], format="mixed", dayfirst=True)
````

````python
df1["Amount"] = df1["Amount"].str.replace("$", "", regex=False)
df1["Amount"] = pd.to_numeric(df1["Amount"])
````

````python
df1["Order_Year"] = df1["Order_Date"].dt.year
df1["Order_Month"] = df1["Order_Date"].dt.month_name()
df1["Order_Day"] = df1["Order_Date"].dt.day
df1["Day_of_Week"] = df1["Order_Date"].dt.day_name()
````

````python
df1["Sales_Category"] = pd.qcut(
    df1["Amount"],
    q=3,
    labels=["Low", "Medium", "High"]
)
````

````python
df1.head()
````

````python
df1.info()
````

### What all manipulations have you done and insights you found?

- Removed all rows containing missing values using the `dropna()` function to ensure a clean dataset.
- Converted the **Order_Date** column to the datetime format for easier time-based analysis.
- Removed the **'$'** symbol from the **Amount** column and converted it to a numeric data type for calculations.
- Created new date-based features such as **Order_Year**, **Order_Month**, **Order_Day**, and **Day_of_Week** to analyze sales trends over time.
- Categorized the **Amount** column into **Low**, **Medium**, and **High** sales groups using `qcut()` for better comparison and visualization.
- Verified the cleaned dataset before proceeding with Exploratory Data Analysis (EDA).
